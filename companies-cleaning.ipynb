```python
import pandas as pd
import numpy as np
```

```python
companies = pd.read_csv("companies/companies.csv")
companies
```

```python
companies.dtypes
```

```python
companies.isnull().sum()
```

```python
rows_with_nulls = companies[companies.isnull().any(axis=1)]
```

```python
rows_with_nulls
```

Because the amount of rows with null values is considerably lower than 20% of the whole dataset we could remove all these rows still remain with a dataset from which we could extract useful information

```python
companies_cleaned = companies.dropna()
```

```python
companies_cleaned
```

```python
companies_cleaned = companies_cleaned.drop_duplicates()
companies_cleaned
```

```python
companies_cleaned.to_csv('companies/companies_cleaned.csv', index=False)
```
