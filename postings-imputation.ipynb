```python
import pandas as pd
import numpy as np
```

```python
posting_df = pd.read_csv("postings.csv")
```

```python
posting_df.isnull().sum()
```

```python
# Impute categorical columns with the mode
categorical_cols = ['company_name', 'pay_period', 'formatted_work_type', 'remote_allowed', 'currency', 'compensation_type']
for col in categorical_cols:
    posting_df[col].fillna(posting_df[col].mode()[0], inplace=True)
```

```python
posting_df.isnull().sum()
```

```python
columns_to_keep = ['med_salary', 'min_salary', 'max_salary']
columns_to_drop = [col for col in posting_df.select_dtypes(include=[np.number]).columns if col not in columns_to_keep]
posting_df.drop(columns=columns_to_drop, inplace=True)
```

```python
numerical_cols = ['max_salary', 'min_salary', 'med_salary']
for col in numerical_cols:
    posting_df[col].fillna(posting_df[col].median(), inplace=True)
```

```python
# Impute categorical columns with the mode
categorical_cols = ['company_name', 'pay_period', 'formatted_work_type','currency', 'compensation_type']
for col in categorical_cols:
    posting_df[col].fillna(posting_df[col].mode()[0], inplace=True)
```

```python
posting_df.isnull().sum()
```

```python
posting_df.drop(columns=["application_url", "formatted_experience_level", "skills_desc","posting_domain", "job_posting_url"], inplace=True)
```
